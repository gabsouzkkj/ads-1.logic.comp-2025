1. Escreva um programa que leia um n√∫mero e verifique se √© ou n√£o um n√∫mero primo.

```python
num = int(input("Digite um n√∫mero: "))

if num <= 1:
    print(f"{num} n√£o √© n√∫mero primo!")
else:
    primo = True
    for n in range(2, num):
        if num % n == 0:
            primo = False
            break
    
    if primo:
        print(f'{num} √© um n√∫mero primo!')
    else:
        print(f'{num} n√£o √© um n√∫mero primo!')
```

2. Escreva um programa para controlar uma pequena m√°quina registradora. Voc√™ deve solicitar ao usu√°rio que digite o c√≥digo do produto e a quantidade desejada. Utilize a tabela de c√≥digos a seguir para obter o pre√ßo de cada produto.


```python
print('Ol√°, seja bem-vindo a M√°quina Registradora! Solicite o produto em base ao c√≥digo dele:')
print('- 1 : R$ 0.50 \n- 2 : R$ 1.00 \n- 3 : R$ 4.00 \n- 4 : R$ 7.00 \n- 5 : R$ 8.00')
print('\nPara finalizar a compra, digite 0!')

armazemCompras = 0
limiteProduto = 5
x = 0

while True:
    x += 1
    codigoProduto = int(input('Insira o c√≥digo do produto > '))

    if codigoProduto == 0:
            break
    
    match codigoProduto:
         case 1:
            preco = 0.50
         case 2:
            preco = 1.00
         case 3:
            preco = 4.00
         case 4:
            preco = 7.00
         case 5:
            preco = 8.00
         case _:
            print('Insira o n√∫mero do c√≥digo do produto corretamente. Tente novamente!')

    quantidadeProduto = int(input('Insira a quantidade de produtos > '))

    subtotal = preco * quantidadeProduto 
    armazemCompras += subtotal

    print(f"Compra {x} - Subtotal adicionado: R$ {subtotal:.2f}")
print(f"Total da compra: R$ {armazemCompras:.2f}")
```
