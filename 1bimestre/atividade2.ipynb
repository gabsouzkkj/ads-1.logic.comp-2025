3. Crie um programa que some todos os números de 1 a 100 e exiba o resultado no final (apenas o total). (1 ponto)

5. Escreva um programa que permita registrar o valor de compras até que o cliente finalize a compra digitando 0. No final, exiba o total da compra. (1 ponto)


```python
totalCompra = 0

while True:
    compra = float(input('Registre aqui o valor de compras e no final será somado ao todo (0 para finalizar) > '))
    if compra == 0:
        break
    totalCompra += compra

print(f'A soma de todos os produtos foram de {totalCompra} reais!')

```

8. Faça um algoritmo que leia 4 notas de um aluno. (2 pontos)

O programa deve:
- Solicitar o nome do aluno.
- Calcular a média aritmética das 4 notas.
- Validar a nota:
- `Aprovado` caso maior ou igual que 7
- `Recuperação` caso entre 4 e 6.9
- `Reprovado` caso menor que 4
- `Nota inválida` caso a nota calculada não esteja entre 0 e 10

```python
x = 1
soma = 0

while x <= 4:
    nota = int(input(f'Digite a {x} nota:'))
    if nota < 0 or nota > 10:
        print("Nota inválida! Digite um valor entre 0 e 10.")
         
    soma += nota
    x += 1

if nota >= 7:
    print(f'Aluno aprovado! Média das notas foram: {soma /4}')
elif nota >= 4 and nota <= 6.9:
    print(f'Aluno está em recuperação! A média das notas é {soma / 4}')
else:
    print(f'Aluno reprovado! A média das notas é {soma / 4}')
```
