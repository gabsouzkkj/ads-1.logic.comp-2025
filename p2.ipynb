1. Escreva um programa que leia um número e verifique se é ou não um número primo.

```python
numero = int(input('Insira um número para verificação de número primo > '))

if numero <= 1:
    print('Não, ele não é um número primo!')
elif numero == 2 or numero == 3:
    print('Ele é um número primo!')
elif numero % 2 == 0 or numero % 3 == 0:
    print('Não, ele não é um número primo!')
else:
    print('Ele é um número primo!')
```

2. Escreva um programa para controlar uma pequena máquina registradora. Você deve solicitar ao usuário que digite o código do produto e a quantidade desejada. Utilize a tabela de códigos a seguir para obter o preço de cada produto.


```python
print('Olá, seja bem-vindo a Máquina Registradora! Solicite o produto em base ao código dele:')
print('- 1 : R$ 0.50 \n- 2 : R$ 1.00 \n- 3 : R$ 4.00 \n- 4 : R$ 7.00 \n- 5 : R$ 8.00')
print('\nPara finalizar a compra, digite 0!')

armazemCompras = 0
x = 0

while True:   
    x += 1
    codigoProduto = int(input('Insira o código do produto > '))

    if codigoProduto == 0:
            break
    
    match codigoProduto:
         case 1:
            preco = 0.50
         case 2:
            preco = 1.00
         case 3:
            preco = 4.00
         case 4:
            preco = 7.00
         case 5:
            preco = 8.00
         case _:
            print('Insira o número do código do produto corretamente. Tente novamente!')

    quantidadeProduto = int(input('Insira a quantidade de produtos > '))

    subtotal = preco * quantidadeProduto 
    armazemCompras += subtotal

    print(f"Compra {x} - Subtotal adicionado: R$ {subtotal:.2f}")

print(f"Total da compra: R$ {armazemCompras:.2f}")
```

3. Você deve desenvolver um programa para acompanhar a pontuação de um time no Campeonato Brasileiro de Futebol rodada por rodada.

```python
inserirTime = input('Insira o nome do time no Campeonato Brasileiro de Futebol > ')

rodada = 1
vitoria = 0
pontos = 0 
empate = 0
derrota = 0

while True:
    print(f'\n = RODADA {rodada} =')
    print('Adicione aqui a pontuação em base:')
    print('- 1. Vitória (+3p) \n- 2. Empate (+1p) \n- 3. Derrota (0p) \n- 0. Para encerrar o programa')
    escolha = int(input('Insira o número para sua escolha > '))

    if escolha == 0:
        break

    match escolha:
        case 1:
            vitoria += 3
            pontos += 1
            print('\nDessa vez foi vitória! +3 pontos')

            placar = input('Insira quanto foi o jogo > ')
            print(f'O placar do jogo foi {placar}!')
            
        case 2:
            empate += 1
            pontos += 1
            print('\nEmpate. +1 pontos')

            placar = input('Insira quanto foi o jogo > ')
            print(f'O placar do jogo foi {placar}!')
        case 3:
            derrota += 1
            print('\nInfelizmente foi derrota.. 0 pontos')

            placar = input('Insira quanto foi o jogo > ')
            print(f'O placar do jogo foi {placar}!')
        case _: 
            print('Insira o número da sua escolha corretamente, releia a lista!')

    print(f"Pontos atuais: {pontos}")
    rodada += 1

if pontos > 20:
   classificado = 'Libertadores'
elif pontos > 15:
  classificado = 'Sul Americana'
else:
   classificado = 'Rebaixamento'

print(f'\n= RESULTADO DA RODADA {rodada} =')
print(f'Time: {inserirTime} - Pontuação: {pontos}')
print(f'- Vitórias: {vitoria} \n- Empate: {empate} \n- Derrota: {derrota} \n- Status de classificação: {classificado}')       
```
